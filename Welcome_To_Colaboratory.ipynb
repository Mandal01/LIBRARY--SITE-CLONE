```
import math

# Basic math operations
print(math.sqrt(25))
print(math.pow(2, 3))

# Trigonometric functions
print(math.sin(math.radians(30)))
print(math.cos(math.radians(45)))

# Constants
print(math.pi)
```

```
import numpy as np

# Creating arrays
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.arange(0, 10, 2)

# Basic operations
sum_arr = arr1 + arr2
dot_product = np.dot(arr1, arr2)

# Mathematical functions
mean_value = np.mean(arr1)
max_value = np.max(arr2)

# Reshaping arrays
reshaped_arr = arr1.reshape(5, 1)

print(sum_arr)
print(dot_product)
print(mean_value)
print(max_value)
print(reshaped_arr)
```

```
import matplotlib.pyplot as plt

x = np.linspace(0, 15, 140)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a scatter plot using Seaborn
sns.scatterplot(x=x, y=y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot')

# Show the plot
plt.show()
```

```
import numpy as np
from scipy.linalg import solve

# Coefficient matrix
A = np.array([[2, 1], [1, 3]])

# Right-hand side vector
b = np.array([5, 8])

# Solve the system of equations Ax = b
x = solve(A, b)

print("Solution:", x)
```
